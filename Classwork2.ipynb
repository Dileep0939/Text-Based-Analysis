# Text Tokenization and Stopword Removal

```python
# Tokenizing the text in the 'text' column using a custom function 'tokenize'
import regex as re
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)

# Testing the tokenize function on a sample text
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
print(testtokens)
```

```python
# Downloading the NLTK stopwords dataset
import nltk
nltk.download('stopwords')

# Creating a set of English stopwords
stopwords = set(nltk.corpus.stopwords.words('english'))

# Displaying the set of stopwords
print(stopwords)
```

```python
# Removing stopwords from tokens
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]

# Testing the remove_stop function
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

```python
# Defining a preprocessing pipeline
pipeline = [str.lower, tokenize, remove_stop]

# Applying the preprocessing pipeline to the 'text' column and creating 'tokens' column
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

# Token Frequency Analysis

```python
# Importing Counter to count token frequencies
from collections import Counter

# Counting token frequencies in a sample text
tokens = tokenize("She likes my cats and my cats like my sofa.")

testcount= Counter(tokens)
print(testcount)
```

```python
# Updating token frequencies with more tokens
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

```python
# Counting token frequencies across the entire DataFrame
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
# Creating a DataFrame from the Counter and sorting by frequency
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
freq_df.describe()
```

```python
# Plotting the top 15 words by frequency
ax = freq_df.head(15).plot(kind='barh', width=0.95, figsize=(12,8))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```

# WordCloud Visualization

```python
# Importing WordCloud and matplotlib for visualization
from wordcloud import WordCloud
from matplotlib import pyplot as plt

# Extracting text for WordCloud for the year 2015 and country USA
text = df.query("year==2015 and country=='USA'")['text'].values[0]

# Generating and displaying WordCloud
plt.figure(figsize=(14, 7)) ###
wc = WordCloud(max_words=100, stopwords=stopwords)
wc.generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```

```python
# Generating and displaying WordCloud from token frequencies
wc = WordCloud(background_color='white', width = 300, height=300, max_words=100, margin=2)
wc.fit_words(freq_df['freq'].to_dict() )
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
